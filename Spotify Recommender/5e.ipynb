```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

df = pd.read_csv('/content/a.csv') 
```

```python
df = df.drop(columns=['Radiojavan_play', 'energy'])
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

newdf = df.select_dtypes(include=numerics)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
 
df_scaled = scaler.fit_transform(newdf.to_numpy())
df_scaled = pd.DataFrame(df_scaled, columns=list(newdf.columns))
 
print("Scaled Dataset Using MinMaxScaler")
df_scaled.head()

df = df_scaled
```

```python
def graph(col_name):
  df_mean = np.mean(df[col_name])
  df_std = np.std(df[col_name])

  pdf = stats.norm.pdf(df[col_name].sort_values(), df_mean, df_std)
  plt.plot(df[col_name].sort_values(), pdf)
  plt.show()


  ax = df[col_name].plot.hist(bins=10, alpha=0.5)
```

```python
graph('Radiojavan_likes')
```
