```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
    
df = pd.read_csv('/content/a.csv') 
```

```python
df_mean = np.mean(df['Radiojavan_play'])
df_std = np.std(df['Radiojavan_play'])

pdf = stats.norm.pdf(df['Radiojavan_play'].sort_values(), df_mean, df_std)
plt.plot(df['Radiojavan_play'].sort_values(), pdf)
plt.grid(True, alpha=0.3, linestyle="--")
plt.show()


ax = df['Radiojavan_play'].plot.hist(bins=100, alpha=0.5)
```
