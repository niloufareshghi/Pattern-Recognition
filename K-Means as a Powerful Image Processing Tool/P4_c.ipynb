```python
import pandas as pd
import numpy as np
from skimage import io
from sklearn.cluster import KMeans
import os
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('trump_tie_3.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
io.imshow(img)
```

```python
img_size = os.path.getsize('/content/trump_tie_3.jpg')
print(img_size)
```

```python
data = img.reshape((img.shape[0] * img.shape[1], img.shape[2]))
print(data)
```

For reduction to 50% of the original image size;

size wanted = 72,806

```python
kmeans = KMeans(n_clusters = 7) 
y_pred = kmeans.fit_predict(data)
centers = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
compressed_img = centers[labels]
compressed_img = np.clip(compressed_img.astype('uint8'), 0, 255)
out1 = compressed_img.reshape((img.shape))
io.imshow(out1)
io.imsave('/content/out1.png', out1)
```

```python
size = os.path.getsize('/content/out1.png')
print(size)
```
