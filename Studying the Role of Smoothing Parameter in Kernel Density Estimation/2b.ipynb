```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def p1(x):
  if 0<x<1:
    return x
  elif 1<x<2:
    return 2-x
  else:
    return 0
```

```python
def p2(x):
  return x * (math.exp(-1 * x * x / 2))
```

```python
t = np.arange(-10,10, 0.0001)
plt.plot(t, list(map(p1, t)), c='r')
plt.plot(t, list(map(p2, t)), c='b')
plt.show()
```

```python
def sampling(f,N):
    sample = np.random.random(N)
    y=[]
    sum = 0
    for x in sample:
      tmp = f(x);
      sum += tmp;
      y.append(tmp)
    y = y / sum;
    return sample, y, np.random.choice(sample, p=y, size=N)
```

```python
s11, o11 , p11 = sampling(p1, 10)
s12, o12 , p12 = sampling(p1, 100)
s13, o13 , p13 = sampling(p1, 1000)
```

```python
plt.show(plt.hist(p11))
plt.show(plt.hist(p12))
plt.show(plt.hist(p13))
```

```python
s21, o21 , p21 = sampling(p2, 10)
s22, o22 , p22 = sampling(p2, 100)
s23, o23 , p23 = sampling(p2, 1000)
```

```python
plt.show(plt.hist(p21))
plt.show(plt.hist(p22))
plt.show(plt.hist(p23))
```
