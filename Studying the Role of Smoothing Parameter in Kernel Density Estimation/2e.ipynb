```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def p1(x):
  if 0<x<1:
    return x
  elif 1<x<2:
    return 2-x
  else:
    return 0
```

```python
def p2(x):
  return x * (math.exp(-1 * x * x / 2))
```

```python
t = np.arange(-10,10, 0.0001)
plt.plot(t, list(map(p1, t)), c='r')
plt.plot(t, list(map(p2, t)), c='b')
plt.show()
```

```python
def sampling(f,N):
    sample = np.random.random(N)
    y=[]
    sum = 0
    for x in sample:
      tmp = f(x);
      sum += tmp;
      y.append(tmp)
    y = y / sum;
    return sample, y, np.random.choice(sample, p=y, size=N)
```

```python
s11, o11 , p11 = sampling(p1, 10)
s12, o12 , p12 = sampling(p1, 100)
s13, o13 , p13 = sampling(p1, 1000)
```

```python
def sample_std(N, d):
  std = 0
  tmp = 0
  for i in range(N):
    avg = sum(d) / N
    tmp += (d[i] - avg)**2 / (N - 1)
  std = math.sqrt(tmp / (N - 1))
  return std
```

```python
std_p1_1 = sample_std(10, o11)
std_p1_2 = sample_std(100, o12)
std_p1_3 = sample_std(1000, o13)
print(std_p1_1)
print(std_p1_2)
print(std_p1_3)
```

```python
hstar_10_p1 = 1.06 * std_p1_1 * (10)**(-1.5)
hstar_100_p1 = 1.06 * std_p1_2 * (100)**(-1.5)
hstar_1000_p1 = 1.06 * std_p1_3 * (1000)**(-1.5)
print(hstar_10_p1)
print(hstar_100_p1)
print(hstar_1000_p1)
```

```python
from sklearn.neighbors import KernelDensity
x_test = np.linspace(-10, 10, 1000)[:, np.newaxis]
model = KernelDensity(kernel='gaussian', bandwidth=hstar_10_p1/3)
model.fit(o11.reshape(-1,1))
log_dens = model.score_samples(x_test)
plt.fill(x_test, np.exp(log_dens), c='cyan')
plt.plot(x_test, list(map(p1, x_test)), c='r')
plt.show()
```
