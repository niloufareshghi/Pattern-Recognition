```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
import seaborn as sns
import csv
```

```python
wine = pd.read_csv('wine_quality.csv', delimiter=';')
```

```python
wine
```

```python
X = wine.drop("quality", axis=1)
X = X.values
y = wine["quality"]
y = y.values
```

```python
data_points = [
    np.array([7.5, 0.9, 0.26, 2.3, 0.054, 19, 13, 0.99708, 3.78, 0.55, 9.7]),
    np.array([5.4, 0.78, 0.17, 3.2, 0.084, 11, 58, 0.9987, 2.94, 0.83, 11.8]),
    np.array([8.2, 0.56, 0.46, 1.7, 0.069, 25, 15, 0.997, 3.39, 0.65, 12.5]),
    np.array([6.0, 0.7, 0.01, 4.6, 0.093, 6, 104, 0.99746, 3.12, 0.52, 10.5]),
    np.array([10.8, 0.43, 0.31, 2.5, 0.105, 35, 31, 1.0001, 3.22, 0.48, 11.1]),
]
```

```python
# 1-NN
for i, data in enumerate(data_points):  
  distances = np.linalg.norm(X - data, axis=1)
  k = 1
  nn_ids = distances.argsort()[:k]
  nn_bodies = y[nn_ids]
  nn_bodies
  prediction = nn_bodies.mean()
  print("predicted quality for {} is {}".format(i, prediction))
```

```python
# 3-NN
for i, data in enumerate(data_points):  
  distances = np.linalg.norm(X - data, axis=1)
  k = 3
  nn_ids = distances.argsort()[:k]
  nn_bodies = y[nn_ids]
  nn_bodies
  prediction = nn_bodies.mean()
  print("predicted quality for {} is {}".format(i, prediction))
```
