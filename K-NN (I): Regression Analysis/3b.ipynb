```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
import seaborn as sns
```

```python
animal = pd.read_fwf('animals_weight.txt')
```

```python
X = animal.drop("Body", axis=1)
X = X.values
y = animal["Body"]
y = y.values
```

```python
data_points = [
    np.array([53.298]),
    np.array([1247.122]),
    np.array([0.583]),
    np.array([4.859]),
    np.array([0.041]),
]
```

```python
# 1-NN
pred_data = []
for data in data_points:  
  distances = np.linalg.norm(X - data, axis=1)
  k = 1
  nn_ids = distances.argsort()[:k]
  nn_bodies = y[nn_ids]
  nn_bodies
  prediction = nn_bodies.mean()
  pred_data.append([data.item(), prediction])
  print("predicted body weight for {} is {}".format(data.item(), prediction))
```

```python
df = pd.DataFrame(pred_data, columns=['Brain', 'Body'])
fig, axs = plt.subplots(ncols=2)
sns.regplot(x="Brain", y="Body", data=df, ci=False, ax = axs[0])
sns.regplot(x="Brain", y="Body", data=animal, ci=False, ax = axs[1])
```
